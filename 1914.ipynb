```python
qt = int(input())

for _ in range(qt):
    # Read player info
    parts = input().split()
    p1_name = parts[0]
    p1_choice = parts[1]
    p2_name = parts[2]
    p2_choice = parts[3]

    # Read numbers
    n, m = map(int, input().split())

    total = n + m

    if total % 2 == 0:
        # even → PAR wins
        if p1_choice == "PAR":
            print(p1_name)
        else:
            print(p2_name)
    else:
        # odd → IMPAR wins
        if p1_choice == "IMPAR":
            print(p1_name)
        else:
            print(p2_name)
```
