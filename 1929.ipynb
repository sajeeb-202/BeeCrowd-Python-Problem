```python
def is_triangle(a, b, c):
    return a + b > c and a + c > b and b + c > a

A, B, C, D = map(int, input().split())

if (
    is_triangle(A, B, C) or
    is_triangle(A, B, D) or
    is_triangle(A, C, D) or
    is_triangle(B, C, D)
):
    print("S")
else:
    print("N")
```
